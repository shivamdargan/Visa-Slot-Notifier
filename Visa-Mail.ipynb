## GMAIL API

```python
from __future__ import print_function
import webbrowser
traveldocsurl = "https://atlas.my.salesforce-sites.com/"

import os.path

from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import datetime
import time
from playsound import playsound
import random


# If modifying these scopes, delete the file token.json.
SCOPES = ['https://www.googleapis.com/auth/gmail.readonly']
faces = ["( ͡° ͜ʖ ͡°)","(▀̿Ĺ̯▀̿ ̿)","¯\_(ツ)_/¯","ಠ_ಠ","(͡ ͡° ͜ つ ͡͡°)"]
flag = 0

def main():
    """Shows basic usage of the Gmail API.
    Lists the user's Gmail labels.
    """
    creds = None
    # The file token.json stores the user's access and refresh tokens, and is
    # created automatically when the authorization flow completes for the first
    # time.
    if os.path.exists('newToken.json'):
        creds = Credentials.from_authorized_user_file('newToken.json', SCOPES)
    # If there are no (valid) credentials available, let the user log in.
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                'CLIENT_SECRET.JSON', SCOPES)
            creds = flow.run_local_server(port=0)
        # Save the credentials for the next run
        with open('newToken.json', 'w') as token:
            token.write(creds.to_json())

    try:
        service = build('gmail', 'v1', credentials=creds)
        while(True):
            try: 
                # GET MESSAGES LIST
                print(datetime.datetime.now(), "LOOKING FOR NEW SLOTS... "+random.choice(faces))
                messagesList = service.users().messages().list(userId='me', q = "from: alerts@checkvisaslots.com is:unread").execute()
                # IF SLOTS ARE AVAILABLE
                if(messagesList['resultSizeEstimate'] > 0 ):
                    # GETTING MESSAGE SUBJECT OF THE EMAIL
                    messageData = service.users().messages().get(userId='me', id = messagesList['messages'][0]['id'] ).execute()
                    for i in messageData['payload']['headers']:
                        if(i['name'] == 'Subject'):
                            print("SLOTS FOUND" , i['value'])
                            flag=1
                    # IF SUBJECT OF THE EMAIL IS NOT FOUND
                    if(flag == 0):
                        print("Error Finding Subject !")
                    ##CALL MOBILE PHONE FUNCTIONS
                    callShivam()
                    callAbhi()
                    #OPEN BROWSER 
                    webbrowser.open(traveldocsurl)
                    # PLAYING ALERT SOUND WHEN SLOTS ARE FOUND 
                    playsound("Red Alert Ringtone-tones6.com.mp3");
                    time.sleep(10)
                else:
                    print(datetime.datetime.now(), "Still looking... "+random.choice(faces))
                    time.sleep(5)
            except Exception as e:
                print(datetime.datetime.now(), "API Unresponsive", e)
                time.sleep(20)
        
    except HttpError as error:
        # TODO(developer) - Handle errors from gmail API.
        print(f'An error occurred: {error}')


if __name__ == '__main__':
    main()
```
